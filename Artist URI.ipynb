```python
artist_uris = []
index = 0
for i in artists:
    if len(sp.search(q = i, limit = 1, type = "artist")['artists']['items']) > 0:
        artist_uris.append(sp.search(q = i, limit = 1, type = "artist")['artists']['items'][0]['id'])
    else:
        artist_uris.append("NaN")
    index += 1
```
